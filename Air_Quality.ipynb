```python
pollutants = df_chang.columns[5:11]
```

```python
pollutants1 = df_guan.columns[5:11]
```

```python
monthly_avg_changping = df_chang.groupby("month")[pollutants].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_avg_changping.index, y=monthly_avg_changping.mean(axis=1), marker="o",linewidth=2.5, color="blue")

plt.title("Grafik Polutan di Changping",fontsize=13,fontweight="bold")
plt.xlabel("Bulan",fontsize=13,fontweight="bold")
plt.ylabel("Rata-rata",fontsize=13,fontweight="bold")
plt.xticks(range(1, 13))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
monthly_avg_guanyuan = df_guan.groupby("month")[pollutants1].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_avg_guanyuan.index, y=monthly_avg_guanyuan.mean(axis=1), marker="o",linewidth=2.5, color="blue")

plt.title("Grafik Polutan di Guanyuan",fontsize=13,fontweight="bold")
plt.xlabel("Bulan",fontsize=13,fontweight="bold")
plt.ylabel("Rata-rata",fontsize=13,fontweight="bold")
plt.xticks(range(1, 13))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Menggunakan gaya seaborn agar lebih rapi
sns.set_style("whitegrid")

# Menghitung rata-rata polutan per bulan di kedua stasiun
monthly_avg_changping = df_chang.groupby("month")[pollutants].mean()
monthly_avg_guanyuan = df_guan.groupby("month")[pollutants1].mean()

# Warna dan marker untuk setiap polutan
colors = sns.color_palette("tab10", len(pollutants))
markers = ['o', 's', '^', 'D', 'v', 'P']

# Membuat subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for i, pollutant in enumerate(pollutants):
    ax = axes[i]
    ax.plot(monthly_avg_changping.index, monthly_avg_changping[pollutant],
            marker=markers[i], linestyle='-', color=colors[i], label=f"Changping - {pollutant}")
    ax.plot(monthly_avg_guanyuan.index, monthly_avg_guanyuan[pollutant],
            marker=markers[i], linestyle='dashed', color=colors[i], label=f"Guanyuan - {pollutant}")
    ax.set_title(f"{pollutant}", fontsize=12, fontweight='bold')
    ax.set_xlabel("Bulan")
    ax.set_ylabel("Konsentrasi (µg/m³)")
    ax.set_xticks(range(1, 13))
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.legend()

# Konfigurasi layout
plt.suptitle("Rata-rata Polutan Udara Per Bulan", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Menggunakan gaya seaborn agar lebih rapi
sns.set_style("whitegrid")

# Menghitung rata-rata polutan per bulan di kedua stasiun
monthly_avg_changping = df_chang.groupby("day")[pollutants].mean()
monthly_avg_guanyuan = df_guan.groupby("day")[pollutants1].mean()

# Warna dan marker untuk setiap polutan
colors = sns.color_palette("tab10", len(pollutants))
markers = ['o', 's', '^', 'D', 'v', 'P']

# Membuat subplots
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
axes = axes.ravel()

for i, pollutant in enumerate(pollutants):
    ax = axes[i]
    ax.plot(monthly_avg_changping.index, monthly_avg_changping[pollutant],
            marker=markers[i], linestyle='-', color=colors[i], label=f"Changping - {pollutant}")
    ax.plot(monthly_avg_guanyuan.index, monthly_avg_guanyuan[pollutant],
            marker=markers[i], linestyle='dashed', color=colors[i], label=f"Guanyuan - {pollutant}")
    ax.set_title(f"{pollutant}", fontsize=12, fontweight='bold')
    ax.set_xlabel("Hari")
    ax.set_ylabel("Konsentrasi (µg/m³)")
    ax.set_xticks(range(1, 32))
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.legend()

# Konfigurasi layout
plt.suptitle("Rata-rata Polutan Udara Per Hari", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Menggunakan gaya seaborn agar lebih rapi
sns.set_style("whitegrid")

# Menghitung rata-rata polutan per bulan di kedua stasiun
monthly_avg_changping = df_chang.groupby("month")[pollutants].mean()
monthly_avg_guanyuan = df_guan.groupby("month")[pollutants].mean()

# Menghitung rata-rata keseluruhan polutan untuk setiap kota
mean_pollutants_changping = monthly_avg_changping.mean()
mean_pollutants_guanyuan = monthly_avg_guanyuan.mean()

# Membuat bar chart untuk membandingkan rata-rata polutan
x = np.arange(len(pollutants))  # Lokasi untuk setiap polutan
width = 0.35  # Lebar bar

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, mean_pollutants_changping, width, label='Changping', color='blue', alpha=0.7)
rects2 = ax.bar(x + width/2, mean_pollutants_guanyuan, width, label='Guanyuan', color='red', alpha=0.7)

# Menambahkan label di atas batang
for rect in rects1:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, f'{height:.1f}', ha='center', va='bottom', fontsize=10)

for rect in rects2:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, f'{height:.1f}', ha='center', va='bottom', fontsize=10)

# Konfigurasi plot
ax.set_title("Perbandingan Rata-rata Polutan antara Changping dan Guanyuan", fontsize=14, fontweight='bold')
ax.set_xlabel("Polutan", fontsize=12)
ax.set_ylabel("Konsentrasi (µg/m³)", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(pollutants)
ax.legend()
ax.grid(axis='y', linestyle="--", alpha=0.5)

# Tampilkan plot
plt.tight_layout()
plt.show()
```
